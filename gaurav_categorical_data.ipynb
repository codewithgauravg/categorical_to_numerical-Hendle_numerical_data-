```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('adult.csv')
df.sample(5)
```

```python
df.shape
```

```python
df.info()
```

```python
df.describe
```

```python
# Clean Column Names
df.columns = df.columns.str.replace('-', '_')
df.columns
```

```python
#Identify Categorical + Numeric Columns
categorical_cols = df.select_dtypes(include='object').columns.tolist()
numeric_cols = df.select_dtypes(include=['int64','float64']).columns.tolist()

categorical_cols, numeric_cols
```

```python
#Handle Missing Categorical Values
df[categorical_cols] = df[categorical_cols].replace("?", pd.NA)

# Impute missing values (most frequent category)
for col in categorical_cols:
    df[col].fillna(df[col].mode()[0], inplace=True)
```

```python
#Label Encoding (each column separately)
from sklearn.preprocessing import LabelEncoder

df_label = df.copy()
le_dict = {}

for col in categorical_cols:
    le = LabelEncoder()
    df_label[col] = le.fit_transform(df_label[col])
    le_dict[col] = le

df_label.head()
```

```python
#Frequency Encoding
df_freq = df.copy()

for col in categorical_cols:
    freq = df_freq[col].value_counts()
    df_freq[col] = df_freq[col].map(freq)

df_freq.head()
```

```python
# Target Encoding
df_target = df.copy()
target_col_name = "income"

# Convert target column to numerical (0 or 1) for encoding
df_target[target_col_name] = df_target[target_col_name].apply(lambda x: 1 if x == ">50K" else 0)

# Identify categorical features to be target encoded (excluding the target column itself)
categorical_features_for_target_encoding = [col for col in categorical_cols if col != target_col_name]

for col in categorical_features_for_target_encoding:
    # Calculate the mean of the target for each category
    means = df_target.groupby(col)[target_col_name].mean()
    df_target[col] = df_target[col].map(means)

df_target.head()


from sklearn.model_selection import train_test_split
import pandas as pd

# Prepare features for one-hot encoding
features_to_onehot = [col for col in categorical_cols if col != 'income']

# Combine numeric columns and one-hot encoded categorical features for X
X = pd.get_dummies(df[numeric_cols + features_to_onehot], columns=features_to_onehot, drop_first=True)

y = df['income'].apply(lambda x: 1 if x == ">50K" else 0)

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)
```

```python
# Handle Imbalanced Dataset
from imblearn.over_sampling import SMOTE

sm = SMOTE(random_state=42)
X_train_sm, y_train_sm = sm.fit_resample(X_train, y_train)

print("Before SMOTE:", y_train.value_counts())
print("After SMOTE:", y_train_sm.value_counts())
```

```python
# Train Random Forest Classifier

from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report

rf = RandomForestClassifier(n_estimators=200, random_state=42)
rf.fit(X_train_sm, y_train_sm)

y_pred = rf.predict(X_test)
```

```python
# Model Evaluation
print("Accuracy:", accuracy_score(y_test, y_pred))
print("\nClassification Report:\n", classification_report(y_test, y_pred))
```
